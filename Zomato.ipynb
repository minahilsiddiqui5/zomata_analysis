```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

## Observation

5. when rating is between 2.0 to 2.4 --->poor

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y="Rating count", data = rating,hue='Rating color', palette = ['white','red','orange','yellow','green', 'green']);
```

observation:
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python

# Create a figure and set its size
plt.figure(figsize=(12, 6))

# Calculate the total count of ratings
total_count = len(df)

# Create a countplot with the 'Rating text' 
order = df['Rating text'].value_counts().index

# Plot the countplot
sns.countplot(data=df, x='Rating text', color='tab:blue', order=order)

# Add percentage labels to the bars
for i, count in enumerate(df['Rating text'].value_counts()):
    pct_string = f'{100 * count / total_count:.1f}%'
    plt.text(i, count + 1, pct_string, va='center', ha='center', fontsize=12)

# Set labels and title
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')

# Show the plot
plt.show()
```
